# **3. Data Cleaning**

**3.1 Identifying Outliers using Z-Score**

```python
from scipy import stats

df = df[~(df["Gender"] == "Other")]
# Calculate Z-scores of each value in the DataFrame
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]


```

**3.2 Remove Outliers**

```python
# Remove rows with outliers based on Z-score
df_no_outliers = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
```

# **Survey**

**What is the distribution of Gender in the dataset?**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Get value counts and percentages for Gender
gender_counts = df['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']
gender_counts['Percentage'] = (gender_counts['Count'] / len(df['Gender'])) * 100

# Create an interactive bar chart with Plotly
fig = px.bar(gender_counts, 
             x='Gender', 
             y='Count', 
             text='Percentage',  # Show percentage directly on the bars
             hover_data={'Count': True, 'Percentage': ':.2f'},  # Show count and percentage on hover
             title='Gender Distribution',
             labels={'Gender':'Gender', 'Count':'Number of Employees'},
             color='Count',  # Add color based on count
             color_continuous_scale='Reds')  # Use a red color theme

# Update layout for better presentation
fig.update_layout(
    xaxis_title='Gender',
    yaxis_title='Count',
    title_x=0.5,  # Center title
    template='plotly_white'
)

# Display chart
fig.show()
```

**How do different numerical variables like Age, Experience, Stress, and Workload correlate with each other?**

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Create the heatmap with a redder color scale
fig = ff.create_annotated_heatmap(
    z=corr_matrix.values,
    x=list(corr_matrix.columns),
    y=list(corr_matrix.index),
    annotation_text=corr_matrix.round(2).values,  # Rounded correlation values
    colorscale=[[0, '#ffffff'], [0.5, '#f88c8c'], [1, '#c70039']],  # Red color scale
    zmin=-1, zmax=1,  # Set range for correlation values
    colorbar=dict(title='Correlation'),
)

# Update layout for better presentation
fig.update_layout(
    title='Correlation Heatmap',
    xaxis_title='Variables',
    yaxis_title='Variables',
    title_x=0.5,  # Center title
    template='plotly_white',
)

# Display the interactive heatmap
fig.show()
```

**How do numerical variables in the dataset (e.g., Age, Experience, Stress, JobSatisfaction) correlate with each other?**

```python
import plotly.graph_objects as go

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Create heatmap
fig = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    colorscale='reds',  # Use a red color scale
    colorbar=dict(title='Correlation'),
    zmin=-1,  # Minimum value for color scale
    zmax=1,   # Maximum value for color scale
    text=corr_matrix.values,  # Display correlation values
    texttemplate='%{text:.2f}',  # Format text values to 2 decimal places
    textfont=dict(size=12, color='black')  # Font size and color for text
))

# Update layout for better presentation
fig.update_layout(
    title='Correlation Matrix of Numerical Variables',
    xaxis_title='Variables',
    yaxis_title='Variables',
    xaxis=dict(tickangle=-45),  # Rotate x-axis labels for readability
    yaxis=dict(tickangle=0),    # Align y-axis labels
    template='plotly_white',
    title_x=0.5  # Center title
)

# Display the interactive heatmap
fig.show()
```
